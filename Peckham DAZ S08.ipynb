```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv(r"C:\Users\wderi\Downloads\brazilian_amazon_fires_1999_2019.csv")
df.head()
```

```python
df["firespots"].mean()
```

```python
print("Missing values in each column:")
print(df.isnull().sum())
```

```python
state_fires = df.groupby("state")["firespots"].sum().reset_index()
state_fires = state_fires.sort_values(by="firespots", ascending=False)
plt.figure(figsize=(10, 5))
plt.bar(state_fires["state"], state_fires["firespots"], color="blue")
plt.title("Trend of Fires in the Brazilian Amazin by State (1999-2019)")
plt.xlabel("State")
plt.ylabel("Number of Fires")
plt.show()
```
